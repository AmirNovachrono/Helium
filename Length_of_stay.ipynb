```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Ensure datetime conversion for date columns
fullData['admitted_at'] = pd.to_datetime(fullData['admitted_at'], errors='coerce')
fullData['discharged_at'] = pd.to_datetime(fullData['discharged_at'], errors='coerce')
fullData['updated_at'] = pd.to_datetime(fullData['updated_at'], errors='coerce')
fullData['dob'] = pd.to_datetime(fullData['dob'], errors='coerce')

# Drop rows with invalid dates (if necessary)
fullData = fullData.dropna(subset=['admitted_at', 'discharged_at', 'dob'])

# Calculate length of stay (LOS) in days (target)
fullData['length_of_stay'] = (fullData['discharged_at'] - fullData['admitted_at']).dt.days

# Filter rows with valid LOS (non-negative values)
fullData = fullData[fullData['length_of_stay'] >= 0]


# Calculate age from dob and updated_at
fullData['age'] = (fullData['updated_at'] - fullData['dob']).dt.days // 365

# Map facility_type to binary values (0 for clinic, 1 for hospital)
fullData['type'] = fullData['type'].str.lower()
facility_type_mapping = {'hospital': 1, 'clinic': 0}
fullData['facility_type_encoded'] = fullData['type'].map(facility_type_mapping)

# Map visit_type to binary values (0 for out-patient, 1 for in-patient)
fullData['visit_type'] = fullData['visit_type'].str.lower()
visit_type_mapping = {'out-patient': 0, 'in-patient': 1}
fullData['visit_type_encoded'] = fullData['visit_type'].map(visit_type_mapping)

# Calculate length of stay (LOS) in days
fullData['length_of_stay'] = (fullData['discharged_at'] - fullData['admitted_at']).dt.days

# Filter rows with valid LOS (non-negative values)
fullData = fullData[fullData['length_of_stay'] >= 0]

# Count visit_id per institution
visit_count_per_institution = (
    fullData.groupby('institution_id')['visit_id']
    .nunique()
    .reset_index(name='visit_id_count')
)

# Aggregate visit data by date and institution_id
visit_data = (
    fullData.groupby(['updated_at', 'institution_id'])
    .size()
    .reset_index(name='visit_count')
)

# Calculate the unique count of patients (volume) per institution
patient_count_per_institution = (
    fullData.groupby('institution_id')['patient_id']
    .nunique()
    .reset_index(name='patient_count')
)

# Calculate the number of unique states each institution is present in
states_per_institution = (
    fullData.groupby('institution_id')['state']
    .nunique()
    .reset_index(name='state_count')
)

# Calculate the average age per institution
avg_age_per_institution = (
    fullData.groupby('institution_id')['age']
    .mean()
    .reset_index(name='avg_age')
)

# Calculate the total visit count per institution
total_visits_per_institution = (
    fullData.groupby('institution_id')
    .size()
    .reset_index(name='total_visit_count')
)

# Aggregate facility_type_encoded per institution
facility_type_per_institution = (
    fullData.groupby('institution_id')['facility_type_encoded']
    .mean()
    .reset_index(name='facility_type_encoded')
)

# Aggregate visit_type_encoded per institution
visit_type_per_institution = (
    fullData.groupby('institution_id')['visit_type_encoded']
    .mean()
    .reset_index(name='visit_type_encoded')
)

# Merge all features into one dataframe
institution_features = pd.merge(total_visits_per_institution, patient_count_per_institution, on='institution_id', how='left')
institution_features = pd.merge(institution_features, avg_age_per_institution, on='institution_id', how='left')
institution_features = pd.merge(institution_features, states_per_institution, on='institution_id', how='left')
institution_features = pd.merge(institution_features, facility_type_per_institution, on='institution_id', how='left')
institution_features = pd.merge(institution_features, visit_type_per_institution, on='institution_id', how='left')
institution_features = pd.merge(institution_features, visit_count_per_institution, on='institution_id', how='left')

# Merge patient count, state count, average age, facility type, visit type, and visit_id_count into visit_data
visit_data = pd.merge(visit_data, patient_count_per_institution, on='institution_id', how='left')
visit_data = pd.merge(visit_data, states_per_institution, on='institution_id', how='left')
visit_data = pd.merge(visit_data, avg_age_per_institution, on='institution_id', how='left')
visit_data = pd.merge(visit_data, facility_type_per_institution, on='institution_id', how='left')
visit_data = pd.merge(visit_data, visit_type_per_institution, on='institution_id', how='left')
visit_data = pd.merge(visit_data, visit_count_per_institution, on='institution_id', how='left')

# Include LOS as target in visit_data
visit_data['length_of_stay'] = fullData.groupby(['updated_at', 'institution_id'])['length_of_stay'].mean().reset_index(drop=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import LabelEncoder, StandardScaler
import numpy as np

# Encode the 'type' column (hospital = 1, clinic = 0)
encoder = LabelEncoder()
fullData['type_encoded'] = encoder.fit_transform(fullData['type'])

# Calculate length of stay in days
fullData['length_of_stay'] = (fullData['discharged_at'] - fullData['admitted_at']).dt.days
df_merged = fullData.dropna(subset=['length_of_stay'])  # Drop rows with missing length_of_stay

# Aggregate data by date and institution_id
aggregated_data = (
    df_merged.groupby(['inserted_at', 'institution_id'])
    .agg(
        visit_count=('patient_id', 'size'),
        length_of_stay=('length_of_stay', 'mean'),  # Average length of stay
        type_encoded=('type_encoded', 'first')
    )
    .reset_index()
)

# Feature Engineering: Create lag features, rolling averages, and other time-based features
def create_features(df):
    df['day_of_week'] = df['inserted_at'].dt.dayofweek
    df['rolling_mean_7'] = df['visit_count'].rolling(window=7).mean()
    df['rolling_mean_30'] = df['visit_count'].rolling(window=30).mean()
    df['rolling_std_7'] = df['visit_count'].rolling(window=7).std()

    # Fill missing values created by shifting and rolling operations
    df.fillna(0, inplace=True)
    return df

aggregated_data = create_features(aggregated_data)

# Apply StandardScaler to scale features
scaler = StandardScaler()

# List of features to scale (excluding the target variable `length_of_stay`)
features_to_scale = ['day_of_week', 'rolling_mean_7', 'rolling_mean_30', 'rolling_std_7', 
                     'visit_count', 'type_encoded']

# Scale the features
aggregated_data[features_to_scale] = scaler.fit_transform(aggregated_data[features_to_scale])

# Create a function to train and forecast SARIMAX for each institution with exogenous variables
def forecast_length_of_stay(institution_id):
    # Filter data for the institution
    institution_data = aggregated_data[aggregated_data['institution_id'] == institution_id].copy()

    # Set the index to the time column
    institution_data = institution_data.set_index('inserted_at').asfreq('D').fillna(0)

    # Define endogenous (target) variable (length_of_stay)
    endog = institution_data['length_of_stay']

    # Define exogenous variables (scaled features including 'visit_count' and 'type_encoded')
    exog = institution_data[['day_of_week', 'rolling_mean_7', 'rolling_mean_30', 'rolling_std_7',
                             'visit_count', 'type_encoded']]

    # Split into train and test sets (90% train, 10% test)
    train_size = int(len(endog) * 0.9)
    train_endog, test_endog = endog[:train_size], endog[train_size:]
    train_exog, test_exog = exog[:train_size], exog[train_size:]

    # Fit the SARIMAX model with exogenous variables
    model = SARIMAX(train_endog, exog=train_exog, order=(1, 1, 1), seasonal_order=(1, 1, 1, 7))
    model_fit = model.fit(disp=False)

    # Forecast the future length of stay
    forecast = model_fit.forecast(steps=len(test_endog), exog=test_exog)

    # Plot the results
    plt.figure(figsize=(12, 6))
    plt.plot(train_endog, label='Train')
    plt.plot(test_endog, label='Test', color='orange')
    plt.plot(forecast, label='Forecast', color='green')
    plt.title(f'Length of Stay Prediction for Institution {institution_id} (SARIMAX with Exogenous Variables)')
    plt.xlabel('Date')
    plt.ylabel('Length of Stay (days)')
    plt.legend()
    plt.show()

    # Evaluate the model using Mean Absolute Error (MAE) and Root Mean Squared Error (RMSE)
    mae = mean_absolute_error(test_endog, forecast)  # MAE
    rmse = np.sqrt(mean_squared_error(test_endog, forecast))  # RMSE
    print(f"Institution {institution_id} - Mean Absolute Error (MAE): {mae}")
    print(f"Institution {institution_id} - Root Mean Squared Error (RMSE): {rmse}")

# Get a list of unique institution_ids
institution_ids = aggregated_data['institution_id'].unique()

# Forecast length of stay for each institution using SARIMAX
for institution_id in institution_ids[:10]:
    forecast_length_of_stay(institution_id)
```
